1. Write a Python program to reverse a string without using any built-in string reversal functions.



```python
def reverse_string(string):
    reversed_string = ""
    for i in range(len(string) - 1, -1, -1):
        reversed_string += string[i]
    return reversed_string

# Example 
input_string = input("Enter a string: ")
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)
```

2. Implement a function to check if a given string is a palindrome.


```python
def is_palindrome(string):
    # Remove whitespace and convert to lowercase
    string = string.replace(" ", "").lower()
    
    # Check if the string is equal to its reverse
    return string == string[::-1]

# Example 
input_string = input("Enter a string: ")
if is_palindrome(input_string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

5. Write a Python program to find the second largest number in a list.


```python
def find_second_largest(lst):
    if len(lst) < 2:
        return None

    largest = max(lst[0], lst[1])
    second_largest = min(lst[0], lst[1])

    for num in lst[2:]:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest

# Example 
input_list = input("Enter a list of numbers: ").split()
input_list = [int(num) for num in input_list]
second_largest = find_second_largest(input_list)
print("The second largest number is:", second_largest)
```

7. Write a program to calculate the factorial of a given number.


```python
def factorial(n):
    if n < 0:
        return None
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example 
input_number = int(input("Enter a number: "))
factorial_result = factorial(input_number)
if factorial_result is not None:
    print("Factorial:", factorial_result)
else:
    print("Factorial is undefined for negative numbers.")
```

14. Implement a function to calculate the Fibonacci sequence up to a given number of terms.


```python
def fibonacci_sequence(n):
    sequence = [0, 1]
    if n <= 2:
        return sequence[:n]
    while len(sequence) < n:
        next_number = sequence[-1] + sequence[-2]
        sequence.append(next_number)
    return sequence

# Example 
input_terms = int(input("Enter the number of terms: "))
fibonacci_sequence = fibonacci_sequence(input_terms)
print("Fibonacci sequence:", fibonacci_sequence)
```

15. Write a program to find the median of a list of numbers.


```python
def find_median(numbers):
    numbers.sort()
    n = len(numbers)
    if n % 2 == 0:
        mid = n // 2
        median = (numbers[mid - 1] + numbers[mid]) / 2
    else:
        mid = n // 2
        median = numbers[mid]
    return median

# Example
input_list = input("Enter a list of numbers: ").split()
input_list = [float(num) for num in input_list]
median = find_median(input_list)
print("Median:", median)
```

22. Implement a function to find the first non-repeating character in a string.


```python
def first_non_repeating_char(string):
    char_count = {}
    for char in string:
        char_count[char] = char_count.get(char, 0) + 1

    for char in string:
        if char_count[char] == 1:
            return char
    return None

# Example
string = "abracadabra"
first_non_repeating = first_non_repeating_char(string)
if first_non_repeating:
    print("The first non-repeating character is:", first_non_repeating)
else:
    print("No non-repeating character found.")
```

23. Write a program to find the prime factors of a given number.


```python
def prime_factors(n):
    factors = []
    i = 2
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors

# Example
number = 84
factors = prime_factors(number)
print("Prime factors of", number, "are:", factors)
```

25. Write a Python program to merge two sorted lists into a single sorted list.


```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    merged_list.extend(list1[i:])
    merged_list.extend(list2[j:])
    return merged_list

# Example
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged = merge_sorted_lists(list1, list2)
print("Merged sorted list:", merged)
```

26. Implement a function to find the mode of a list of numbers.


```python
from collections import Counter

def find_mode(numbers):
    counts = Counter(numbers)
    max_count = max(counts.values())
    mode = [num for num, count in counts.items() if count == max_count]
    return mode

# Example
numbers = [1, 2, 3, 4, 2, 3, 4, 4, 5]
mode = find_mode(numbers)
print("Mode(s) of the numbers:", mode)
```

30. Implement a function to find the minimum element in a rotated sorted list.


```python
def find_min_element_rotated(arr):
    low = 0
    high = len(arr) - 1

    while low < high:
        mid = low + (high - low) // 2

        if arr[mid] > arr[high]:
            low = mid + 1
        else:
            high = mid

    return arr[low]

# Example
rotated_list = [4, 5, 6, 7, 0, 1, 2]
min_element = find_min_element_rotated(rotated_list)
print("Minimum element in the rotated list:", min_element)
```

34. Implement a function to find the longest common prefix among a list of strings.


```python
def longest_common_prefix(strings):
    if not strings:
        return ""

    prefix = strings[0]
    for string in strings[1:]:
        while string[:len(prefix)] != prefix:
            prefix = prefix[:-1]
            if prefix == "":
                return ""
    return prefix

# Example
string_list = ["flower", "flow", "flight"]
common_prefix = longest_common_prefix(string_list)
print("Longest common prefix:", common_prefix)
```

41. Write a Python program to find the smallest missing positive integer in a list.


```python
def find_smallest_missing_positive_integer(nums):
    if not nums:
        return 1

    n = len(nums)
    contains_one = False

    # Step 1: Mark non-positive numbers and check if 1 is present
    for i in range(n):
        if nums[i] == 1:
            contains_one = True
        if nums[i] <= 0 or nums[i] > n:
            nums[i] = 1

    # If 1 is not present, return 1
    if not contains_one:
        return 1

    # Step 2: Mark visited indices
    for i in range(n):
        idx = abs(nums[i])
        if idx == n:
            nums[0] = -abs(nums[0])
        else:
            nums[idx] = -abs(nums[idx])

    # Step 3: Find the smallest missing positive integer
    for i in range(1, n):
        if nums[i] > 0:
            return i

    if nums[0] > 0:
        return n

    return n + 1

# Example 
num_list = [3, 4, -1, 1]
smallest_missing = find_smallest_missing_positive_integer(num_list)
print("Smallest missing positive integer:", smallest_missing)
```

42. Implement a function to find the longest palindrome substring in a given string.


```python
def longest_palindrome_substring(string):
    n = len(string)
    max_length = 0
    start = 0

    for i in range(n):
        # Check for odd-length palindromes centered at i
        left = right = i
        while left >= 0 and right < n and string[left] == string[right]:
            current_length = right - left + 1
            if current_length > max_length:
                max_length = current_length
                start = left
            left -= 1
            right += 1

        # Check for even-length palindromes centered at i and i+1
        left = i
        right = i + 1
        while left >= 0 and right < n and string[left] == string[right]:
            current_length = right - left + 1
            if current_length > max_length:
                max_length = current_length
                start = left
            left -= 1
            right += 1

    return string[start:start + max_length]

# Example
input_string = "babad"
longest_palindrome = longest_palindrome_substring(input_string)
print("Longest palindrome substring:", longest_palindrome)
```

44. Implement a function to check if a given number is a perfect number.


```python
def is_perfect_number(number):
    if number <= 0:
        return False
    divisors = [1]
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            divisors.append(i)
            divisors.append(number // i)
    return sum(divisors) == number

# Example
num = 28
if is_perfect_number(num):
    print(num, "is a perfect number.")
else:
    print(num, "is not a perfect number.")
```

46. Implement a function to find the first missing positive


```python
def find_first_missing_positive(nums):
    n = len(nums)
    i = 0

    # Place each positive number at its correct index
    while i < n:
        if 0 < nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            nums[nums[i] - 1], nums[i] = nums[i], nums[nums[i] - 1]
        else:
            i += 1

    # Find the first missing positive number
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1

    return n + 1

# Example 
num_list = [3, 4, -1, 1]
first_missing_positive = find_first_missing_positive(num_list)
print("First missing positive:", first_missing_positive)
```
